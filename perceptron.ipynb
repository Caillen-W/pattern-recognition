```python
x = np.array([[1,2],[5,3],[8,5],[3,4],[8,4],[4,4]])
np.matrix(x[0])
```

```python
y = np.transpose(np.array([[1, -1, -1, 1, -1, 1]]))
y
```

```python
J = - sum(np.multiply(y, (x @ w + b)))
J
```

```python
while True: 
    if (np.any(np.multiply(y, (x @ w + b)) <= 0)):
        for i in range(len(y)):
            if (y[i] * (np.matrix(x[i]) * w + b))[0][0] <= 0:
                w = w + r * y[i].item() * np.transpose(np.matrix(x[i]))
                b = b + r * y[i].item()
                print(w, b)
            else:
                continue
    else:
        break
w,b
```
